```
!pip install contextualSpellCheck
```

Carrega o corretor ao pipeline do spaCy

```
import contextualSpellCheck

model_name = "neuralmind/bert-large-portuguese-cased"

# Realizado o download do modelo do BERT e adiciona o contextual spellchecker para o modelo do spaCy
nlp.add_pipe(
    "contextual spellchecker",
    config={
        "model_name": model_name,
        "max_edit_dist": 2,
    },
);
```

Mostra os elementos do pipeline

```
nlp.pipe_names
```

## 4.1 - Exemplos

### 4.1.1 Exemplo 1

```
documento =  "A receita foi de $ 9,4 milões em comparação com o ano anterior de $ 2,7 milões."

doc = nlp(documento)

if doc._.performed_spellCheck == True:
  print("O documento foi corrigido para:")
  print("    ",doc._.outcome_spellCheck)
else:
  print("O documento não tem nada a ser corrigido!")
```
